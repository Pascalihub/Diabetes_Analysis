### Introduction
Analyzing diabetes involves understanding the condition, its types, risk factors, symptoms, diagnosis, treatment, and management. Diabetes is a chronic metabolic disorder characterized by elevated blood sugar levels. There are primarily two types of diabetes: Type 1 and Type 2, with gestational diabetes being another type that occurs during pregnancy. Here's a brief overview:

Types of Diabetes:

Type 1 Diabetes: This is an autoimmune disease where the immune system mistakenly attacks and destroys insulin-producing beta cells in the pancreas. People with Type 1 diabetes require insulin injections for survival.

Type 2 Diabetes: This is the most common form of diabetes, often associated with lifestyle factors like poor diet, lack of exercise, and obesity. In Type 2 diabetes, the body doesn't use insulin properly, leading to insulin resistance.

Gestational Diabetes: Occurs during pregnancy and usually resolves after childbirth. It can increase the risk of both the mother and the child developing Type 2 diabetes later in life.

Risk Factors:

Type 1 Diabetes: Genetics and family history play a significant role.
Type 2 Diabetes: Obesity, physical inactivity, family history, age, and certain ethnic backgrounds are risk factors.
Gestational Diabetes: Family history and obesity are risk factors.
Symptoms:

Frequent urination
Excessive thirst and hunger
Fatigue
Blurred vision
Slow wound healing
Unexplained weight loss (Type 1)
Recurrent infections
Diagnosis:

Fasting blood sugar test
Oral glucose tolerance test
Hemoglobin A1c test
Random blood sugar test
Glycated albumin (GA) test
Treatment:

Type 1 Diabetes: Insulin therapy (multiple daily injections or insulin pump).
Type 2 Diabetes: Lifestyle changes (diet and exercise), oral medications, and sometimes insulin.
Gestational Diabetes: Dietary changes and, in some cases, insulin.
Management:

Regular monitoring of blood sugar levels.
Balanced diet with carbohydrate counting.
Regular physical activity.
Medications or insulin as prescribed.
Monitoring for complications (e.g., eye, kidney, and nerve damage).
Complications:

Long-term uncontrolled diabetes can lead to various complications, including cardiovascular disease, kidney disease, neuropathy, retinopathy, and more.
Prevention:

Type 1 diabetes cannot be prevented.
Type 2 diabetes can often be prevented or delayed through lifestyle changes like maintaining a healthy weight, eating a balanced diet, and regular exercise.
Research and Advances:

Ongoing research explores better treatments, early detection, and prevention strategies.
Continuous glucose monitoring and artificial pancreas systems are among the technological advancements.
It's essential for individuals at risk or diagnosed with diabetes to work closely with healthcare professionals to develop a personalized management plan and lead a healthy life. Diabetes management aims to keep blood sugar levels within a target range to prevent complications and maintain overall health.

## ***Fill missing Values*** 
We will replace the 0-values of the following columns with their median values:

*   Glucose
*   BloodPressure
*   SkinThickness
*   Insulin
*   BMI

```python
# first copy the original dataframe into a new pandas dataframe
df_cp = df.copy()
# now fill the missing values with the median value of the particular columns
df_cp['Glucose'] = df['Glucose'].replace(0, df['Glucose'].median())
df_cp['BloodPressure'] = df['BloodPressure'].replace(0, df['BloodPressure'].median())
df_cp['SkinThickness'] = df['SkinThickness'].replace(0, df['SkinThickness'].median())
df_cp['Insulin'] = df['Insulin'].replace(0, df['Insulin'].median())
df_cp['BMI'] = df['BMI'].replace(0, df['BMI'].median())
```

```python
df_cp.describe().T
```

With this, we replaced all the missing values with the median (of that column) in the Training set and used those *learned* medians to replace the missing values in the Testing set.
